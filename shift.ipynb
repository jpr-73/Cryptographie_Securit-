```python
def shift(msg, n):
    res = ''
    for c in msg: 
        if c.isalpha():
            base = ord('A') if c.isupper() else ord('a')
            nw_c = chr((ord(c) - base + n) % 26 + base)
            res += (nw_c)
        else:
            res += (c)
    return res

print(shift("HELLO", 5))


def xor(msg, key):
    text = msg.encode('utf-8')
    key = key.encode('utf-8')
    print(text)
    print(key)


    res = bytes([t ^ k for t, k in zip(text, key)])
    
    print(res)

    decoded = bytes([res[i] ^ key[i % len(key)] for i in range(len(res))])
    print(f"decoded {decoded.decode('utf-8')}")

    return res

print(xor('abc', 'OP'))

def xor3():
    text = "abcdef"
    key = "HM"
    
    
    text_bytes = text.encode('utf-8')
    key_bytes = key.encode('utf-8')
    
    
    if len(key_bytes) < len(text_bytes):
        
        adjusted_key = (key_bytes * (len(text_bytes) // len(key_bytes) + 1))[:len(text_bytes)]
    else:
        
        adjusted_key = key_bytes[:len(text_bytes)]
    
    
    result_bytes = bytes([text_bytes[i] ^ adjusted_key[i] for i in range(len(text_bytes))])
    
    print(f"Text: {text} → {text_bytes}")
    print(f"Key: {key} → {key_bytes}")
    print(f"Adjusted key: {adjusted_key}")
    print(f"XOR result: {result_bytes}")
    
    return result_bytes



print(xor3())
```
